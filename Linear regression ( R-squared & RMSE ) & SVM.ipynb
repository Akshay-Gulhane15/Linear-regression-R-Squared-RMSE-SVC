#	Building Linear Regression Model to predict the BoxOfficeCollection also RMSE and R-squared value for the model.

```python
#Importing the libraries and importing dataset
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

bollywood= pd.read_csv(r'C:\Users\ASUS\OneDrive\Pictures\bollywood.csv')
```

```python
bollywood.head()
```

```python
bolly_col = ['ReleaseTime','Genre']
```

```python
bollywood.isnull().sum()
bollywood.info()
```

```python
bollywood.isnull().sum()
```

```python

bollywood_final = pd.get_dummies(bollywood, columns=bolly_col, drop_first=True)
```

```python
bollywood_final.head()
```

```python
Y = bollywood_final[['BoxOfficeCollection']]

X = bollywood_final[['Budget','ReleaseTime_HS','ReleaseTime_LW','ReleaseTime_N','Genre_Action','Genre_Action','Genre_Comedy','Genre_Romance','Genre_Thriller','Genre_Thriller']]


#STEP 3 = SPLITING DATA INTO TRAINING 80% & TEST DATA 20%

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size=0.8, random_state= 4321)

len(X_train), len(X_test), len(Y_train), len(Y_test)
```

```python
#: Build a blank object using linear regression function
from sklearn.linear_model import LinearRegression

 

slr = LinearRegression()

 

# Fit the object on training data

 

model = slr.fit(X_train,Y_train)

 

model

print ('The Model has run successfuly')
```

```python
# Finding the slope:(m)
print(model.coef_)

 

# find the intercept or constant:(C)
print(model.intercept_)

 

# FInd the r-sq value of model
r_sq = model.score(X_train, Y_train)
r_sq
```

```python
Y_test.head()
```

```python
X_test.head()
```

```python
# Predict the values on test data using your model

Y_test['Predicted_BoxOfficeCollection'] = model.predict(X_test)
```

```python
Y_test
```

```python
Y_test["Predicted_BoxOfficeCollection"] = round(Y_test["Predicted_BoxOfficeCollection"],2)
```

```python
Y_test.head()
```

```python
#Find the value of RMSE (Root <- mean <- square <- error)


Y_test['error'] = (Y_test['BoxOfficeCollection'] - Y_test['Predicted_BoxOfficeCollection'])

 

Y_test['sq_error'] = (Y_test['BoxOfficeCollection'] - Y_test['Predicted_BoxOfficeCollection']) **2

 

# Find the Mean of Sq-error
mean_sq_err = Y_test['sq_error'].mean()
mean_sq_err

 

# Find the sq-root value
import math

 

RMSE = math.sqrt(mean_sq_err)
RMSE
```

```python
hr['attrition']=hr['attrition'].astype('category')
```

```python

air_col = ['salary','department']
```

```python
#creating dummy variables

hr_final = pd.get_dummies(hr, columns=air_col, drop_first=True)
```

```python
hr_final.tail()
```

```python
hr_final.head(12)
```

```python
# For the attirition Column let us convert this into numerical form by using LabelEncoder

from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
hr_final['attrition'] = le.fit_transform(hr_final['attrition'])
hr_final.head()
```

```python
hr['department'].value_counts()
```

```python
Y = hr_final[['attrition']]

X = hr_final.drop(columns=['employee_id','attrition'])

```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scal = sc.fit_transform(X)
```

```python
X
```

```python
# SPlit the data into train and test
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_scal,Y, test_size=0.2, random_state=7777)

len(X_train), len(X_test), len(Y_train), len(Y_test)
```

```python
#Support Vector Machine (SVM)

# import the model object:
from sklearn.svm import SVC
# Create the model object

svm = SVC()

# Fit the object on train data

model = svm.fit(X_train,Y_train)
model
```

```python
Y_test['predict'] = model.predict(X_test)
Y_test
```

```python
# accuracy score and classification report
from sklearn.metrics import accuracy_score, classification_report

accuracy_score(Y_test['attrition'], Y_test['predict'])
print(classification_report(Y_test['attrition'], Y_test['predict']))
```
